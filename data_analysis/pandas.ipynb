# Now making a graph

```python
import matplotlib.pyplot as plt
plt.subplots(figsize=(10,6))
bars = plt.bar(x="player name", height="goals scored", color="yellow", hatch="//", edgecolor="black", width=0.5, data=df.head(10)) #width for size
#name of country on each bar
for bar, nation in  zip(bars,df.head(10).nationality):
    yval= bar.get_height()
    plt.text(x =bar.get_x(), y =yval+.1, s=nation, fontdict ={"size": 7, "color": "blue"}, #yval+.1 to upward y axis data.
    )
plt.xticks(rotation=45)
plt.title("Top 10 Goals Scored in 2022")
plt.show()
```

```python
#clubs with most player 
df["club"].value_counts().head(10).plot(kind="pie", color="orange", autopct="%.1f%%", startangle=90)  #autopct ->automatic percentage. another % for percentage size
```

```python
#clubs with most goals 
bars =df.groupby("club")["goals scored"].sum().sort_values(ascending=False).head(10).plot(kind="bar")
#higlight the 1st place
bars.patches[0].set_color("orange")
#goals scored on each bar
for bar in bars.patches:
    yval= bar.get_height()
    plt.text(x = bar.get_x() + bar.get_width()/2,
             y =1,
             s =f"{int(yval)}",
             ha="center",
             fontdict={"size": 12, "color": "white", "weight": "bold"},
             #rotation=90, #so that no of goals can be displayed vertically.
    )
```

```python
#And for horizontal graph
#clubs with most goals 
bars =df.groupby("club")["goals scored"].sum().sort_values(ascending=False).head(10).plot(kind="bar")
#higlight the 1st place
bars.patches[0].set_color("orange")
```
