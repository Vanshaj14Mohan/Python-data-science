# Fixing the column names

```python
df.columns.tolist() #to remocve the extra spaces.
```

```python
#1 method
for col in df.columns.tolist(): #To remove extra spaces using for loop but we won't use it much.
    print(col.strip())
```

```python
list(map(lambda col: col.strip(), df.columns.tolist())) #strip
```

```python
#and to make it in lowercase
df.columns = list(map(lambda col: col.strip().lower(), df.columns.tolist())) #strip
```

# value_counts()- frequency of unique values in a column

```python
df["goals scored"].value_counts()
```

```python
import numpy as np
```

```python
df["goals scored"].replace("N.A", np.nan, inplace=True)
df["goals scored"].replace("N,A", np.nan, inplace=True)
```

```python
df["goals scored"].value_counts()
```

```python
df["goals scored"].value_counts().plot(kind ="bar", title="Goals Scored")
```

```python
df["goals scored"].value_counts().plot(kind ="pie", title="Goals Scored")
```
